```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 10 # Radio del capacitor [m]
I = 1 # Corriente [A]
z = np.linspace(0, 0.5, 5) #Distancia del eje [m]
c = 3e8 # Velocidad de la luz [m/s]
```

```python
def Bp(p,I,a,z): #Campo magnético a partir de Biot-Savart
    
    ex = (I*p*(z/a))/(c*a**2) # Campo magnético debido al borde de la placa circular 
    
    return (I*p/(c*a**2))*(1 - z/(a-p))+ ex
```

```python
p = np.linspace(0.1, 9, 100) # Radio de corriente superficial
for i in range(len(z)):
    B = Bp(p,I,a,z[i]) #Campo magnético
    plt.scatter(p,B,s=3, label = f'z = {z[i]}')
    plt.legend()
plt.xlabel(r'$\rho$[m]')
plt.ylabel(r'$B_{\phi}[G]$')
plt.grid()
plt.savefig('fig_pdesplegada.pdf')
```

```python
K = -((1 - p**2/a**2)*I)/(2*np.pi*p)
```

```python
plt.scatter(p, K, s=3, c='k')
plt.xlabel(r'$\rho$[m]')
plt.ylabel(r'K $[A/m^2]$')
plt.grid()
plt.savefig('fig_K.pdf')
```
